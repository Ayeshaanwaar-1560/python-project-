```python
# üìä Task: Explore & Visualize Iris Dataset (Modified Version)
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# üé® Different style for visuals
sns.set_theme(style="ticks", palette="muted")

# 1Ô∏è‚É£ Load dataset
print("üì• Importing Iris dataset...")
df = sns.load_dataset("iris")

# üîÑ Rename columns for uniqueness
df = df.rename(columns={
    "sepal_length": "SepalLen",
    "sepal_width": "SepalWid",
    "petal_length": "PetalLen",
    "petal_width": "PetalWid",
    "species": "FlowerType"
})

# 2Ô∏è‚É£ Dataset overview
print("\n--- Dataset Dimensions ---")
print(f"Total Records: {df.shape[0]} | Total Features: {df.shape[1]}")

print("\n--- Feature Names ---")
print(list(df.columns))

print("\n--- Preview of Data ---")
print(df.sample(5))  # using sample instead of head for variation

# 3Ô∏è‚É£ Summary stats
print("\n--- Dataset Info ---")
df.info()

print("\n--- Numerical Summary ---")
print(df.describe())

# 4Ô∏è‚É£ Data Visualization
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# üîπ Histogram (PetalLen)
sns.histplot(
    data=df,
    x="PetalLen",
    hue="FlowerType",
    bins=20,
    alpha=0.6,
    ax=axes[0, 0]
)
axes[0, 0].set_title("Petal Length Distribution")

# üîπ Scatter Plot (SepalLen vs PetalWid)
sns.scatterplot(
    data=df,
    x="SepalLen",
    y="PetalWid",
    hue="FlowerType",
    style="FlowerType",
    s=90,
    ax=axes[0, 1]
)
axes[0, 1].set_title("Sepal Length vs Petal Width")

# üîπ Boxplot (Sepal Width)
sns.boxplot(
    data=df,
    x="FlowerType",
    y="SepalWid",
    palette="pastel",
    ax=axes[1, 0]
)
axes[1, 0].set_title("Sepal Width per Flower Type")

# üîπ KDE Plot (Petal Width)
sns.kdeplot(
    data=df,
    x="PetalWid",
    hue="FlowerType",
    fill=True,
    common_norm=False,
    alpha=0.4,
    ax=axes[1, 1]
)
axes[1, 1].set_title("Petal Width Density Plot")

plt.tight_layout()
plt.show()

# üîπ Pairplot at the end (different order)
sns.pairplot(df, hue="FlowerType", diag_kind="hist")
plt.suptitle("Pairwise Feature Relationships", y=1.02)
plt.show()

print("\n‚úÖ Data Exploration Completed Successfully!")
```
